```python
import torch
import cv2
import warnings
import pytesseract
import time
warnings.filterwarnings(action="ignore")#to ignore further warnings
```

```python

model_path = r"yolov5/runs/train/exp2/weights/best.pt"
model = torch.hub.load('ultralytics/yolov5', 'custom', path=model_path)
```

```python
def print_dates(image):
    text = pytesseract.image_to_string(image)
    print(text)
```

```python
cap = cv2.VideoCapture(0)
if not cap.isOpened():
    print("Error: Could not open webcam.")
    exit()

while True:

    #time.sleep(1) # 

    # start_time = time.time()


    ret, frame = cap.read()
    if not ret:
        print("Error: Failed to capture image.")
        break

 
    results = model(frame)
    detections = results.xyxy[0]  # gets: xmin, ymin, xmax, ymax, confidence, class
    results.render()
    cv2.imshow('YOLOv5 Real-Time Detection', results.ims[0])

    if detections.shape[0] > 0:
        for bbox in detections:
            xmin, ymin, xmax, ymax = map(int, bbox[:4])
            confidence, class_id = map(float, bbox[4:6])
            if confidence >= 0.70:
                cropped_image = frame[ymin:ymax, xmin:xmax]
                print_dates(cropped_image)
                # end_time = time.time()
                # print(end_time - start_time)

        
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break


    

cap.release()
cv2.destroyAllWindows()
```
