```python
import warnings
warnings.filterwarnings("ignore")
import cv2
import torch
import serial
import time

# Load YOLOv5 model
model = torch.hub.load('ultralytics/yolov5', 'custom', path='best.pt', trust_repo=True)

# Serial setup â€” adjust COM port and baud rate as needed
ser = serial.Serial('COM3', 115200)  # For Windows (e.g., COM3)
# ser = serial.Serial('/dev/ttyUSB0', 115200)  # For Linux/Mac

time.sleep(2)  # Wait for Arduino to reset

# Confidence threshold
```

```python
CONF_THRESHOLD = 0.7

# Start camera
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Get predictions
    results = model(frame)
    df = results.pandas().xyxy[0]
    print(df)
    output = '1'  # Default: no class 2 detected

    for _, row in df.iterrows():
        # print(row['class'])
        
        if row['class'] == 0 and row['confidence'] > 0.5:
            output = '1'
            break
        elif row['class'] == 1 and row['confidence'] > CONF_THRESHOLD:
            output = '0'
            break

    print(output)
    
    # Send to Arduino
    # ser.write(output.encode())  # Send as byte string

    # Show frame
    results.render()
    cv2.imshow('YOLOv5', results.ims[0])

    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
ser.close()
cv2.destroyAllWindows()
```
